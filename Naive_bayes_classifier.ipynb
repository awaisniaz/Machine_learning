Basic Probality Theory

P(A/B)  =  P(B/A)*P(A)/P(B)

```
import numpy as np
import pandas as pd
data = pd.read_csv('https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv')
data
```

```

data1  = data[['Pclass','Survived','Sex','Age','Fare']]
data1
```

```
target = data1.Survived
input = data1.drop('Survived',axis=1)
input
```

```
dummies = pd.get_dummies(input['Sex'])
data3 = input.drop('Sex',axis=1)
dummies
```

```
final_data = pd.concat([data3,dummies],axis = 1)
final_data
```

```
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(final_data,input,test_size = 0.30,random_state=0)
len(x_train)
y_train
```

```
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train,y_train)
prediction = model.predict(x_test)
prediction
```
