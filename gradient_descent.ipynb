```python
import numpy as np
import pandas as pd

def graident_descent(x,y):
  m_curr = b_curr = 0
  iteration = 10000
  learning_rate = 0.001
  n  = len(x)
  for i in range(iteration):
    y_prediction = m_curr * x + b_curr
    cost = 1/2*sum(val ** 2  for val in (y-y_prediction))
    md = -(2/n)*sum(x*(y-y_prediction))
    bd = -(2/n)*sum(y-y_prediction)
    m_curr = m_curr - learning_rate * md
    b_curr = b_curr - learning_rate * bd
    print('m {} b  {} *************** cost {}'.format(m_curr,b_curr,cost))


x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

graident_descent(x,y)
```
